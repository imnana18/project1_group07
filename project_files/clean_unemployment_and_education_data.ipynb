```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load
census_data = Path("../datasets/age_census_data.csv")
```

```python
# Load csv as df
census_df = pd.read_csv(census_data)
```

```python
# View df
census_df.head()
```

```python
# Drop all columns with margin in name
census_df.drop(census_df.filter(regex='Margin').columns, axis=1, inplace=True)
census_df.head()
```

```python
# Rename the columns to be more readable 
renamed_census_df = census_df.rename(columns=lambda x: x.replace('!!60 years and over!!Estimate', '_60+_Estimate'))
renamed_census_df = renamed_census_df.rename(columns=lambda x: x.replace('!!Total!!Estimate', '_Total_Estimate'))
renamed_census_df = renamed_census_df.rename(columns={"Label (Grouping)": "Stat"})
renamed_census_df
```

```python
# Filter out Puerto Rico data
renamed_census_df = renamed_census_df.drop(columns=['Puerto Rico_Total_Estimate', 'Puerto Rico_60+_Estimate'])
renamed_census_df
```

```python
# Drop all columns with 60+ data
renamed_census_df.drop(renamed_census_df.filter(regex='60+').columns, axis=1, inplace=True)
renamed_census_df.head()
```

```python
# Filter out the educational related rows
L = [37,38,39,40,41,42]
education_level = renamed_census_df.loc[L]
education_level
```

```python
# Flip the rows and columns and reset the index
education_level_flipped = education_level.set_index('Stat').transpose()
education_level_flipped
```

```python
# Save .csv output
out_path = Path("../datasets/filtered_state_educational_data.csv")
education_level_flipped.to_csv(out_path)
```

```python
# Get the unemployment data per state, by filtering out the unemployment rows
N = [77,78,79,80,81,82,83,84]
unemployment_data = renamed_census_df.loc[N]
unemployment_data
```

```python
# Flip the rows and columns and reset the index
unemployment_data_flipped = unemployment_data.set_index('Stat').transpose()
unemployment_data_flipped
```

```python
# Save .csv output
out_path = Path("../datasets/filtered_unemployment_data.csv")
unemployment_data_flipped.to_csv(out_path)
```
