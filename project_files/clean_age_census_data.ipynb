```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load
census_data = Path("../datasets/age_census_data.csv")
```

```python
# Load csv as df
census_df = pd.read_csv(census_data)
```

```python
# View df
census_df.head()
```

```python
# Drop all columns with margin in name
census_df.drop(census_df.filter(regex='Margin').columns, axis=1, inplace=True)

# Drop 60+ columns
census_df.drop(census_df.filter(regex='60').columns, axis=1, inplace=True)
census_df.head()
```

```python
# Rename the columns to be more readable 
# renamed_census_df = census_df.rename(columns=lambda x: x.replace('!!60 years and over!!Estimate', '_60+_Estimate'))
renamed_census_df = census_df.rename(columns=lambda x: x.replace('!!Total!!Estimate', ''))
renamed_census_df = renamed_census_df.rename(columns={"Label (Grouping)": "Stat"})
renamed_census_df
```

```python
# # Drop all columns with 60+ data
# renamed_census_df.drop(renamed_census_df.filter(regex='60+').columns, axis=1, inplace=True)
# renamed_census_df.head()
```

```python
# Filter out age and Total Pop rows
age_census_df = renamed_census_df.iloc[0:5]
row_to_delete = [1,2,3]
age_census_df.drop(renamed_census_df.index[row_to_delete], inplace=True)
age_census_df
```

```python
# Flip the rows and columns and reset the index
age_data_flipped = age_census_df.set_index('Stat').transpose()
age_data_flipped
```

```python
# Save .csv output
out_path = Path("../datasets/filtered_age_data.csv")
age_data_flipped.to_csv(out_path)
```
