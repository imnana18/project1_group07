```python
# Rename education attainment related columns ('Less than high school graduate')
education_data_df.rename(columns={education_data_df.columns[3]: 'Less than high school graduate'}, inplace=True)

# Remove the percentage sign and convert to float using pd.to_numeric
education_data_df['Less than high school graduate'] = education_data_df['Less than high school graduate'].str.rstrip('%').astype(float)

# Rename education attainment related columns ('Less than high school graduate')
education_data_df.rename(columns={education_data_df.columns[4]: 'High school graduate, GED, or alternative'}, inplace=True)

# Remove the percentage sign and convert to float using pd.to_numeric
education_data_df['High school graduate, GED, or alternative'] = education_data_df['High school graduate, GED, or alternative'].str.rstrip('%').astype(float)

# Rename education attainment related columns ('Less than high school graduate')
education_data_df.rename(columns={education_data_df.columns[5]: 'Some college or associates degree'}, inplace=True)

# Remove the percentage sign and convert to float using pd.to_numeric
education_data_df['Some college or associates degree'] = education_data_df['Some college or associates degree'].str.rstrip('%').astype(float)

# Rename education attainment related columns ('Less than high school graduate')
education_data_df.rename(columns={education_data_df.columns[6]: 'Bachelors degree or higher'}, inplace=True)

# Remove the percentage sign and convert to float using pd.to_numeric
education_data_df['Bachelors degree or higher'] = education_data_df['Bachelors degree or higher'].str.rstrip('%').astype(float)

# Columns related to educational attainment
education_columns = [
    "State",
    "Less than high school graduate",
    "High school graduate, GED, or alternative",
    "Some college or associates degree",
    "Bachelors degree or higher",
]

# Selecting only the relevant columns
education_data_final_df = education_data_df[education_columns]

# Plotting
education_data_final_df.set_index('State').plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Educational Level per State')
plt.xlabel('State')
plt.ylabel('Population')
plt.show()
```

```python
# File to Load (Education)
vaccination_data = Path("../datasets/cleaned_covid19_vaccinations.csv")
```

```python
# Pull csv as df
vaccination_data_df = pd.read_csv(vaccination_data)
vaccination_data_df.head()
```

```python
# Merge DataFrames
merged_education_vaccination_df = pd.merge(education_data_final_df, vaccination_data_df, on='State', how='inner')
merged_education_vaccination_df.head()
```

```python
# Creating a new DataFrame with selected columns
education_vaccination_df = merged_education_vaccination_df[[
    "State", 
    "Less than high school graduate", 
    "High school graduate, GED, or alternative", 
    "Some college or associates degree",
    "Bachelors degree or higher",
    "Percent of total pop with a completed primary series"]]
education_vaccination_df.head()
```

```python
# Plotting a grouped bar chart
fig, ax = plt.subplots(figsize=(10, 6))

bar_width = 0.2
bar_positions = range(len(education_vaccination_df['State']))

for i, col in enumerate(education_vaccination_df.columns[1:5]):
    ax.bar(
        [pos + i * bar_width for pos in bar_positions],
        education_vaccination_df[col],
        width=bar_width,
        label=col
    )

# Plotting the line chart for 'Percent of total pop with a completed primary series'
ax.plot(education_vaccination_df['Percent of total pop with a completed primary series'], marker='o', linestyle='-', color='black', label='Percent of total pop with a completed primary series')

# Setting labels and title
ax.set_xlabel('State')
ax.set_ylabel('Percentage')
ax.set_title('Comparison of Education Levels with Vaccination Percentage by State')

# Setting x-axis ticks and labels
ax.set_xticks([pos + 1.5 * bar_width for pos in bar_positions])
ax.set_xticklabels(education_vaccination_df['State'])

# Adding legend
ax.legend()

# Display the plot
plt.show()
```

```python
# Plotting a scatter plot with a linear regression line
fig, ax = plt.subplots(figsize=(10, 6))

ax.scatter(education_vaccination_df['Less than high school graduate'], education_vaccination_df['Percent of total pop with a completed primary series'], color='blue', label='Data points')

# Calculating and displaying the linear regression line
slope, intercept, r_value, _, _ = linregress(education_vaccination_df['Less than high school graduate'], education_vaccination_df['Percent of total pop with a completed primary series'])
line = slope * education_vaccination_df['Less than high school graduate'] + intercept
ax.plot(education_vaccination_df['Less than high school graduate'], line, color='red', label='Linear Regression Line')

# Displaying the R-value (correlation coefficient)
ax.text(0.95, 0.95, f'R-value: {r_value:.2f}', transform=ax.transAxes, fontsize=12, ha='right', va='top')

# Setting labels and title
ax.set_xlabel('Less than high school graduate (%)')
ax.set_ylabel('Percent of total pop with a completed primary series (%)')
ax.set_title('Less Than High School Graduate vs Vaccination Rate')

# Display the plot
plt.show()
```

```python
# Analysis: 
# The r-value of -0.25 indicates a negative correlation between 
# 'Less Than High School Graduate vs Vaccination Rate'. 
# As the percentage of people with less than a high school education increases, 
# the vaccination rate tends to decrease slightly, but the relationship is low and weak.
```
