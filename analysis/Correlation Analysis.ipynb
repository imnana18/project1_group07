```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt

# File to Load (Age)
age_data = Path("../datasets/filtered_age_data.csv")
```

```python
# Pull csv as df
age_data_df = pd.read_csv(age_data)
age_data_df.head()
```

```python
# File to Load (Education)
education_data = Path("../datasets/filtered_state_educational_data.csv")
```

```python
# Pull csv as df
education_data_df = pd.read_csv(education_data)
education_data_df.head()
```

```python
# File to Load (Unemployment)
unemployment_data = Path("../datasets/filtered_unemployment_data.csv")
```

```python
# Pull csv as df
unemployment_data_df = pd.read_csv(unemployment_data)
unemployment_data_df.head()
```

```python
# Merge DataFrames
merged_age_education_df = pd.merge(age_data_df, education_data_df, on='Unnamed: 0', how='inner')
merged_all_df = pd.merge(merged_age_education_df, unemployment_data_df, on='Unnamed: 0', how='inner')
merged_all_df.head()
```

```python
# Rename State column
renamed_state_column = merged_all_df.rename(columns={"Unnamed: 0": "State"})
renamed_state_column.head()
```

```python
# Rename rows in the State column
renamed_state_column['State'] = renamed_state_column['State'].replace({'_Total_Estimate': ''}, regex=True)
renamed_state_column.head()
```

```python
# Convert 'Total population' column from string to numeric
renamed_state_column['Total population'] = pd.to_numeric(renamed_state_column['Total population'].str.replace(',', ''))

# Adjust total population by dividing by 1 million for better readability
renamed_state_column['Total Population Adjusted'] = renamed_state_column['Total population'] / 1000000

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(renamed_state_column['State'], renamed_state_column['Total Population Adjusted'], color='skyblue')
plt.title('Total Population by State')
plt.xlabel('State')
plt.ylabel('Total Population')
plt.xticks(rotation=45, ha='right') 

# Show the plot
plt.show()
```

```python
# Get the columns name
renamed_state_column.columns
```

```python
# Get data types
renamed_state_column.dtypes
```

```python
# Rename columns
renamed_state_column.rename(columns={renamed_state_column.columns[4]: 'Population 25 Years and Over'}, inplace=True)
renamed_state_column['Population 25 Years and Over'] = renamed_state_column['Population 25 Years and Over'].str.replace(',', '').astype('int64')

# Adjust total population by dividing by 1 million for better readability
renamed_state_column['Population 25 Years and Over Adjusted'] = renamed_state_column['Population 25 Years and Over'] / 1000000

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(renamed_state_column['State'], renamed_state_column['Population 25 Years and Over'], color='skyblue')
plt.title('Total Population of 25 years and over by State')
plt.xlabel('State')
plt.ylabel('Population 25 years and over')
plt.xticks(rotation=45, ha='right') 

# Show the plot
plt.show()
```

```python
# Convert 'Population 16 years and over' column from string to numeric
renamed_state_column.rename(columns={renamed_state_column.columns[10]: 'Population 16 Years and Over'}, inplace=True)
renamed_state_column['Population 16 Years and Over'] = renamed_state_column['Population 16 Years and Over'].str.replace(',', '').astype('int64')

# Adjust total population by dividing by 1 million for better readability
renamed_state_column['Population 16 Years and Over Adjusted'] = renamed_state_column['Population 16 Years and Over'] / 1000000

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(renamed_state_column['State'], renamed_state_column['Population 16 Years and Over'], color='skyblue')
plt.title('Total Population of 16 years and over by State')
plt.xlabel('State')
plt.ylabel('Population 16 years and over')
plt.xticks(rotation=45, ha='right') 

# Show the plot
plt.show()
```

```python
# Rename media age years 
renamed_state_column.rename(columns={renamed_state_column.columns[2]: 'Median age (years)'}, inplace=True)

# Plotting the bar chart for media age years
plt.figure(figsize=(10, 6))
plt.bar(renamed_state_column['State'], renamed_state_column['Median age (years)'], color='skyblue')
plt.title('Median age (years) by State')
plt.xlabel('State')
plt.ylabel('Median age (years)')
plt.xticks(rotation=45, ha='right') 

# Show the plot
plt.show()
```

```python
# Raname Employed name
renamed_state_column.rename(columns={renamed_state_column.columns[13]: 'Employed'}, inplace=True)

# Raname UnEmployed name
renamed_state_column.rename(columns={renamed_state_column.columns[14]: 'Unemployed'}, inplace=True)

# Remove the percentage sign and convert to float using pd.to_numeric
renamed_state_column['Employed'] = pd.to_numeric(renamed_state_column['Employed'].str.rstrip('%'), errors='coerce')
renamed_state_column['Unemployed'] = pd.to_numeric(renamed_state_column['Unemployed'].str.rstrip('%'), errors='coerce')


# Calculate correlation with Employed
correlation_age_employed = renamed_state_column['Median age (years)'].corr(renamed_state_column['Employed'])
print('Employed Correlation', correlation_age_employed)


# Calculare correlation with Unemployed
correlation_age_unemployed = renamed_state_column['Median age (years)'].corr(renamed_state_column['Unemployed'])
print('Unemployed Correlation', correlation_age_unemployed)
```
